# Realizando consultas

```python
from connect import db
from bson.objectid import ObjectId

resultados = db.libros

#Consultar todo

for doc in resultados.find():
    print(doc)
```

```python
# Consulta por autor

for doc in resultados.find({'Autor':'Mario Vargas Llosa'}):
    print(doc['titulo'], doc['Anio'])
```

```python
# Consulta muestre libros mayores al año 1960
for doc in resultados.find({'Anio':{'$gt':1960}}):
    print(doc['titulo'], doc['Anio'])
```

```python
# Listas libros de ciencia ficcion
for doc in resultados.find({'generos':'ciencia ficcion'}):
    print(doc['titulo'], doc['generos'])
```

```python
# libros con stock mayor a 2

for doc in resultados.find({'Stock': 
                            {'$gt': 2}
                            }):
    print(doc['titulo'], doc['Anio'], 'Stock', doc['Stock'], 'Unidades')
```
