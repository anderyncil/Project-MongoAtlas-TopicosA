# Actualizando Documents and Collections

```python
from connect import db
from pymongo import UpdateOne

modificar = db.libros

# Actualizar el stock de un libro
res = modificar.update_one(
    {'titulo':'La ciudad Verde'},
    {'$set': {'Stock': 4}}
)
print('Cambios', res.matched_count, 'Total', res.modified_count)
```

```python
for doc in modificar.find({'titulo':'La ciudad Verde'}):
    print(doc['titulo'], 'Hay', doc['Stock'], 'libros')
```

```python
# Actualizacion usando update many
res2 = modificar.update_many(
    {},
    {'$set': {
         'disponible': True
         }
    }     
)
print('Cambios', res2.matched_count, 'Total', res2.modified_count)
```

```python
for doc in modificar.find({'disponible':True}):
    print('Disponible', doc['disponible'])
```

```python
# Bulk write
operations = [
    UpdateOne({'titulo': 'A'}, {'$set': {'Stock': 5}}),

]
modificar.bulk_write(operations)
```
